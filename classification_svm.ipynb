# SVM Classifier

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("C:\\Users\\sathv\\Downloads\\data (2).csv")
```

```python
df.head(10)
```

```python
x =df.drop(['id','diagnosis'],axis=1)
```

```python
y = df.diagnosis.values
```

```python
x
```

```python
y
```

```python
#train_test_split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25)
```

```python
x_train
```

```python
y_train
```

```python
#feature scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train_sc = sc.fit_transform(x_train)
x_test_sc = sc.transform(x_test)
```

```python
x_train_sc
```

```python
x_test_sc
```

```python
#model building - SVM
from sklearn.svm import SVC
```

```python
model_svm = SVC()
model_svm.fit(x_train_sc,y_train)
```

```python
y_pred_svm = model_svm.predict(x_test_sc)
```

```python
y_pred_svm
```

```python
y_test
```

mostly its the same lets check for the accuracy

```python
# accuracy score
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred_svm)
```

```python
print(round(accuracy_score(y_test,y_pred_svm)*100,2))
```

its 95.8 % of accuracy in svm classification
