lets start with the train and test split as we have segregated the x and y values so here the training data will be trained to the model and testing will be tested by the trained model

```python
#Train & Test Split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25)
```

Feature Scaling - 2 types - Normalization(mean-max scaler) and Standardization(Standard scaler)

```python
#Feature Scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train_sc = sc.fit_transform(x_train)
x_test_sc = sc.transform(x_test)
```

fit_transform - used in training data 
- Fit → learns the parameters from the training set (for example, a scaler learns the mean and standard deviation).
- Transform → applies those parameters to scale the training data.

transform - used on test data
- It doesn’t learn anything new; it simply applies the parameters that were already learned from the training set. This ensures consistency and prevents data leakage.

## MODEL BUILDING - KNN

```python
 # Model Building
from sklearn.neighbors import KNeighborsClassifier
# Intiating the classifer
model_knn = KNeighborsClassifier()
# KNeighborsclassifier() - takes the default value of n_neighbors as 5, and all the different parameters as default values.we can change the default values of the parameters as per our requirement.
# we have intiated the classifer,now we will fit the model on the training data(scalerdata for x).
model_knn.fit(x_train_sc,y_train)
```

the classifier/model is trained with the training data
now lets check the trained model with the test data - prediction

"In supervised learning, we scale the input features 'X' because many algorithms are distance-based or sensitive to feature magnitude. Scaling ensures that no single feature dominates due to its larger numeric range. However, we don’t scale the target variable 'y' because it represents the labels we’re predicting. Scaling 'y' would change the meaning of the classes or regression targets, which would mislead the model. So the rule of thumb is: scale your features, not your labels.”


```python
y_pred_knn = model_knn.predict(x_test_sc)
```

lets check the predicted values

just compare the predicted values with the original values so compare the model how much of the % is predicted correctly.

```python
# Classification metrics to check how the model is performing
# Accuracy score
from sklearn.metrics import accuracy_score
print(round(accuracy_score(y_test,y_pred_knn)*100,2))
```

normally accuracy_score gives result in decimal point so in this stage we have rounded it to 2 decimal points and *100 to get the value in % form.
