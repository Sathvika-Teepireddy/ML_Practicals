```python
#splitting the data into training and testing data
x = df.drop(["id", "diagnosis"], axis=1)
```

```python
y = df.diagnosis.values
```

```python
x
```

```python
#Train & Test Split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25)
```

```python
len(x_train)
```

```python
len(x_test)
```

```python
#Feature Scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train_sc = sc.fit_transform(x_train)
x_test_sc = sc.transform(x_test)
```

```python
x_train_sc
```

```python
x_test_sc
```

# Model Building - decision tree classification

```python
from sklearn.tree import DecisionTreeClassifier
#initiate the classifier
model_dtc = DecisionTreeClassifier()
# passing the training data to the model
model_dtc.fit(x_train_sc,y_train)
```

```python
y_pred_dtc = model_dtc.predict(x_test_sc)
```

```python
y_pred_dtc
```

```python
y_test
```

almost same in comparsion let us compare them through classification metrics - accuracy_score

```python
from sklearn.metrics import accuracy_score
print(round(accuracy_score(y_test,y_pred_dtc)*100,2))
```
